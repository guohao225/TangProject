```python
import json
with open("./word_seg.txt", 'r', encoding='utf8') as fp:
    seg_data = json.load(fp)
```

```python
seg_data
```

```python
all_word = []
for key in seg_data:
    all_word+=seg_data[key]
```

```python
all_word
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tk = Tokenizer()
tk.fit_on_texts(all_word)
```

```python
tk.word_index
```

```python
tk.num_words
```

```python
vocab = [key for key in tk.word_index]
vocab
```

```python
import pandas
pandas.DataFrame({'words':vocab}).to_csv('./segVocab', index=None, header=None)
```
